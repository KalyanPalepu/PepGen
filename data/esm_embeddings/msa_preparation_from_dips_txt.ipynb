```python
from tqdm import tqdm
import pandas as pd
import pickle
from collections import defaultdict
```

```python
data_input = open('pdb_seqres.txt', 'r')
lines = data_input.readlines()
```

```python
complete_list = []
for idx, line in enumerate(tqdm(lines)):
    if idx %2 == 0:
        list_i = [line[1:7], lines[idx + 1][:-1]]
        complete_list.append(list_i)
    
```

```python
df2 = pd.DataFrame(complete_list, columns =[ 'pdbchain_id', 'sequence'])
```

```python
seq_list = list(df2["sequence"])
seq_set = set(seq_list)
unique_seq_list = list(seq_set)
```

```python
d = defaultdict(list)
for k,v in zip(df2.sequence, df2.pdbchain_id):
    d[k].append(v)
    
seq_to_chains_dict = dict(d)
```

```python
hash_table = {}
for vallist in seq_to_chains_dict.values():
    hash_table[vallist[0]] = vallist
```

```python
df_final = df2.drop_duplicates(subset=["sequence"])
```

```python
df_final.to_csv('non_duplicate_seqs_2.csv')
with open('MSA_id_to_all_ids_2.pkl', 'wb') as f:
    pickle.dump(hash_table, f)
```
